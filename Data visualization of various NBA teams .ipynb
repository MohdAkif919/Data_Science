```python
# imported required modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("nba.csv") #reading csv file
pd.set_option("max_rows",None) #showing all records
```

```python
data
```

```python
#created group of all team
group = data.groupby("Team") 
```

```python
#showing information of each player of their respective team
for team, value in group:
    display(team)
    display(value)
    value.plot.bar("Name","Salary")
    plt.xticks(rotation="vertical")
    plt.show()
```
